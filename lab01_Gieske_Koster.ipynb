### Notation

$\mathbf{X}$ is the $M \times T$ data matrix, containing $M$ measurements at $T$ time steps.

$\mathbf{S}$ is the $S \times T$ source matrix, containing $S$ source signal values at $T$ time steps. We will assume $S = M$.

$\mathbf{A}$ is the mixing matrix. We have $\mathbf{X} = \mathbf{A S}$.

$\mathbf{W}$ is the matrix we aim to learn. It is the inverse of $\mathbf{A}$, up to indeterminacies (scaling and permutation of sources).

$\phi$ is an elementwise non-linearity or activation function, typically applied to elements of $\mathbf{W X}$.

In the following assignments, you can make use of the signal generators listed below.



```python
%pylab inline
import itertools

# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
    return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp

def sine_wave(x, period=0.2, amp=1.0, phase=0.):
    return np.sin((x / period - phase) * 2 * np.pi) * amp

def square_wave(x, period=0.2, amp=1.0, phase=0.):
    return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) * 2 * amp

def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp

def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix if shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X):
    """
    Plot the signals contained in the rows of X.
    """
    figure()
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
```

### 1.1 Make mixtures (5 points)
Write a function `make_mixtures(S, A)' that takes a matrix of source signals $\mathbf{S}$ and a mixing matrix $\mathbf{A}$, and generates mixed signals $\mathbf{X}$.

```python
def make_mixtures(S,A):
    """Mixes source signals with matrix
    
    Parameters:
        S -- Matrix of source signals
        A -- Mixing matrix
    
    Returns:
        X -- Matrix of mixed signals
    """
    X = np.dot(A,S)
    return X

#Create mixing matrix A
A = random_nonsingular_matrix(d=5)
#Create mixed signal matrix X
X = make_mixtures(S, A)
```

### 1.4 Whitening (15 points)
Some ICA algorithms can only learn from whitened data. Write a method `whiten(X)` that takes a $M \times T$ data matrix $\mathbf{X}$ (where $M$ is the dimensionality and $T$ the number of examples) and returns a whitened matrix. If you forgot what whitening is or how to compute it, various good sources are available online, such as http://courses.media.mit.edu/2010fall/mas622j/whiten.pdf

```python
def whiten(X):
    # Get covariance matrix of data
    cov = np.cov(X)
    # Get eigenvalues and eigenvectors of covariance matrix
    eigvalues, E = np.linalg.eigh(cov)
    D = np.diag(1. / np.sqrt(eigvalues))
    X_white = np.dot(np.dot(np.dot(E, D), E.T),X)
    return X_white

W_x = whiten(X)
```

### 1.7 Independent Component Analysis (25 points)
Implement the covariant ICA algorithm as described in MacKay. Write a function `ICA(X, activation_function, learning_rate)`, that returns the demixing matrix $\mathbf{W}$. The input `activation_function` should accept a function such as `lambda a: -tanh(a)`. Update the gradient in batch mode, averaging the gradients over the whole dataset for each update. Try to make it efficient, i.e. use matrix operations instead of loops where possible (loops are slow in interpreted languages such as python and matlab, whereas matrix operations are internally computed using fast C code).

```python
def ICA(X, activation_function, learning_rate, threshold = 0.05):
    print "Running ICA (might take a while).."
    #Create random nonsingular (invertible) matrix
    W = random_nonsingular_matrix(X.shape[0])
    shape_W = W.shape[0]
    # Initialize iteration counter
    it = 0;
    # Initialize difference
    diff = 1;
    # Continue ascent until difference is lower than treshold
    while diff > threshold:
        # Reset delta_W
        delta_W = zeros((shape_W,shape_W))
        # Put through linear mapping
        a= np.dot(W,X)
        #Put a through non-linear mapping
        z = activation_function(a)  
        x_hat = np.dot(W.T, a)
        # Trick to speed up: multiply delta W by T to later divide by T
        delta_W = W*X.shape[1] + np.dot(z,x_hat.T)
        # Update W
        W = W+ learning_rate*(delta_W/X.shape[1])
        
        # Calculate highest difference in delta W
        diff = np.amax(abs(delta_W))
        it +=1
        # Uncomment to print iteration number to track progress
        #if it %1000 == 0:
        #    print it

        # Check for convergence
        if diff < threshold:
            print "DONE: total iterations: %i" %(it)
    return W
```

### 1.10 Excess Kurtosis (20 points)
The (excess) kurtosis is a measure of 'peakedness' of a distribution. It is defined as
$$
\verb+Kurt+[X] = \frac{\mu_4}{\sigma^4} - 3 = \frac{\operatorname{E}[(X-{\mu})^4]}{(\operatorname{E}[(X-{\mu})^2])^2} - 3
$$
Here, $\mu_4$ is known as the fourth moment about the mean, and $\sigma$ is the standard deviation.
The '-3' term is introduced so that a Gaussian random variable has 0 excess kurtosis.
We will now try to understand the performance of the various activation functions by considering the kurtosis of the corresponding priors, and comparing those to the empirical kurtosis of our data.

First, compute analytically the kurtosis of the four priors that you derived from the activation functions before. To do this, you will need the normalizing constant of the distribution, which you can either obtain analytically (good practice!), using computer algebra software (e.g. Sage) or by numerical integration (see scipy.integrate).

Now use the `scipy.stats.kurtosis` function, with the `fisher` option set to `True`, to compute the empirical kurtosis of the dummy signals and the real audio signals. Can you use this data to explain the performance of the various activation functions?

As you can see in the plots for the probability distributions in 1.3, all of them have $\mu=0$. Following integrations are solved using wolframalpha.

Expectations:
$E[a^4] = \int\limits_{-\infty}^\infty p(a)a^4da\\
E[a^2] = \int\limits_{-\infty}^\infty p(a)a^2da$

- activation function 0, uses \frac{1}{\pi} as normalization constant C
\begin{align}
p(a) &= \frac{1}{cosh(a)}\\
E[a^4] &= C \int\limits_{-\infty}^\infty \frac{1}{cosh(a)}a^4da\\
&= \frac{1}{\pi} \frac{5\pi^5}{16} = \frac{5\pi^4}{16}\\
E[a^2] &= C \int\limits_{-\infty}^\infty \frac{1}{cosh(a)}a^2da \\
&= \frac{1}{\pi}\frac{\pi^3}{4} = \frac{\pi^2}{4}\\
\verb+Kurt+[X] &= \frac{\frac{5\pi^4}{16}}{(\frac{\pi^2}{4})^2}- 3 = 2
\end{align}

- activation function a1, uses $\frac{1}{\sqrt{2\pi \text{e}}}$ as normalization constant C
\begin{align}
p(a) &= \text{e}^{-\frac{a^2}{2}} \text{cosh}(a)\\
E[a^4] &= C \int\limits_{-\infty}^\infty \text{e}^{-\frac{a^2}{2}} \text{cosh}(a)a^4da\\
&= \frac{1}{\sqrt{2\pi \text{e}}}10\sqrt(2\text{e}\pi)
&= 10\\
E[a^2] &= C \int\limits_{-\infty}^\infty \text{e}^{-\frac{a^2}{2}} \text{cosh}(a)a^2da \\
&= \frac{1}{\sqrt{2\pi \text{e}}} 2\sqrt(2\text{e}\pi)
&= 2\\
\verb+Kurt+[X] &= \frac{10}{2^2} -3 = 0.5
\end{align}


- activation function a2, uses $\frac{\sqrt(2)}{\Gamma(\frac{1}{4})}$ as normalization constant C
\begin{align}
p(a) &=  \text{e}^{-\frac{1}{4}a^4} \\
E[a^4] &= C \int\limits_{-\infty}^\infty \text{e}^{-\frac{1}{4}a^4}a^4da \\
&=\frac{\sqrt(2)}{\Gamma(\frac{1}{4})} \frac{\Gamma(\frac{1}{4})}{\sqrt{2}}\\
&= 1\\
E[a^2] &= C \int\limits_{-\infty}^\infty \text{e}^{-\frac{1}{4}a^4}a^2da \\
&=\frac{\sqrt(2)}{\Gamma(\frac{1}{4})} \sqrt{2}\Gamma(\frac{3}{4})\\
&\approx 0.68
\verb+Kurt+[X] &\approx \frac{1}{0.68^2}-3 \approx -0.84
\end{align}

- activation function a3, uses $\frac{200\sqrt{5}}{3\pi}$ as normalization constant C
\begin{align}
p(a) &=  \frac{1}{(a^2+5)^3}\\
E[a^4] &= C \int\limits_{-\infty}^\infty \frac{1}{(a^2+5)^3}a^4da \\
&= \frac{200\sqrt{5}}{3\pi}\frac{3\pi}{5\sqrt{5}} = 25\\
E[a^2] &= C \int\limits_{-\infty}^\infty \frac{1}{(a^2+5)^3}a^2da\\
&= \frac{200\sqrt{5}}{3\pi} \frac{\pi}{40\sqrt{5}} = \frac{5}{3}\\
\verb+Kurt+[X] &= \frac{40}{(\frac{5}{3})^2}- 3 = 6
\end{align}
